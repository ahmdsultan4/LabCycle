```python
#Q1
import prettytable
def noofpair(m):
  if m==1 or m==2:
    return m
  a, b, c = 1, 2, 3
  while c<=m:
    a, b = b, a + b
    c+=1
  return b

table = prettytable.PrettyTable()
table.field_names = ["Number Of Moths", "Number Of Pairs"]
table.align["Number Of Pairs"] = "l"

m=int(input("Enter the number of months: "))

i=1
while(i<=m):
  table.add_row([i,(" "+str(noofpair(i)))])
  i+=1

print(table)
```

```python
#Q2
def strsplit(s):
  strl=[]
  for i in s.split():
    strl.append(int(i))
  return strl
def rotbyk(s):
  k=int(input("Enter the number of rotations: "))
  k=k%len(s)
  return s[-k:]+s[:-k]
def function(n):
  list2=[]
  for i in n:
    list2.append((i**2)-i)
  return list2
s=input("Enter the numbers: ")
strl=strsplit(s)
strt=tuple(strl)
uniql=list(set(strt))
list2=function(strl)
print("String as list: ",strl)
print("Rotated List: ",rotbyk(strl))
print("String as a Tuple: ",strt)
print("Unique elements: ",uniql)
print("Edited List(x^2-x): ",list2)
print("Sorted List: ",sorted(strl))
print("Sorted Edited List: ",sorted(list2))
print("Sorted Full List: ",sorted(strl+list2))
```

```python
#Q3
import requests
import json
response = requests.get('https://drive.google.com/uc?id=1N2GAGRRKQeL8GDHTcomqF7biZdnfyUVL')
data = response.json()
"""
#Data to check the output- Not Part Of The Code
data=[
      {"sepalLength": 5.1, "sepalWidth": 3.5, "petalLength": 1.4, "petalWidth": 0.2, "species": "setosa"},
      {"sepalLength": 4.9, "sepalWidth": 3.0, "petalLength": 1.4, "petalWidth": 0.2, "species": "setosa"},
      {"sepalLength": 7.0, "sepalWidth": 3.2, "petalLength": 4.7, "petalWidth": 1.4, "species": "versicolor"},
      {"sepalLength": 6.4, "sepalWidth": 3.2, "petalLength": 4.5, "petalWidth": 1.5, "species": "versicolor"},
      {"sepalLength": 6.3, "sepalWidth": 3.3, "petalLength": 6.0, "petalWidth": 2.5, "species": "virginica"},
      {"sepalLength": 5.8, "sepalWidth": 2.7, "petalLength": 5.1, "petalWidth": 1.9, "species": "virginica"}
]
"""
def listtodict(l):
  ddict={}
  for i in range(len(l)):
    ddict[i+1]=l[i]
  return ddict

def printsetosa(datadict):
  for i in datadict:
    if datadict[i]["species"]=="setosa":
      print(datadict[i])

def minpetal(datadict):
  mindict={}
  mindict["setosa"]=1000000
  mindict["versicolor"]=1000000
  mindict["virginica"]=1000000
  for i in datadict:
    species=datadict[i]["species"]
    if datadict[i]["petalLength"]*datadict[i]["petalWidth"]<mindict[species]:
      mindict[species]=datadict[i]["petalLength"]*datadict[i]["petalWidth"]
  return mindict

def maxsepal(datadict):
  maxdict={}
  maxdict["setosa"]=0
  maxdict["versicolor"]=0
  maxdict["virginica"]=0
  for i in datadict:
    species=datadict[i]["species"]
    if datadict[i]["sepalLength"]*datadict[i]["sepalWidth"]>maxdict[species]:
      maxdict[species]=datadict[i]["sepalLength"]*datadict[i]["sepalWidth"]
  return maxdict

def totalarea(l):
  area=(l["sepalLength"]*l["sepalWidth"]+l["petalLength"]*l["petalWidth"])
  return area

def sortbytotalarea(datadict):
  return dict(sorted(datadict.items(), key=lambda item: totalarea(item[1])))

print("Data as a list: \n",data)

datadict=listtodict(data)
print("\n\nData as a dictionary: \n", datadict)

print("\n\nSetosa Flowers: \n")
printsetosa(datadict)

print("\n\nMinimum Petal Area: \n")
for key, value in minpetal(datadict).items():
    print(f"{key} : {value}")
print("\n\nMaximum Sepal Area: \n")
for key, value in maxsepal(datadict).items():
    print(f"{key} : {value}")

print("\n\nSorted by Total Area: \n")
for key,value in sortbytotalarea(datadict).items():
    print(f"{value}")
```

```python
#Q4
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

url='https://drive.google.com/uc?id=1nbqCk9BhOXj5hxfO8RO6wc1iJADOoiXD'
data=pd.read_csv(url)

def interpret_correlation(correlation):
    if correlation > 0.7:
        return "strong positive correlation"
    elif correlation < -0.7:
        return "strong negative correlation"
    elif 0.3 < correlation <= 0.7:
        return "moderate positive correlation"
    elif -0.7 < correlation < -0.3:
        return "moderate negative correlation"
    else:
        return "weak correlation"


print("\n\n1.Data Exploration:")

print("\n  a.1)Size: ",data.shape)
print("\n  a.2)Features available: ",list(data.columns))

print(f'\n  b)Duplicate entries: {data.duplicated().sum()}')
data.drop_duplicates(inplace=True)

print("\n  c) Visualization the distribution of the target variable")
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Wine')
plt.title('Distribution of Wine Classes')
plt.xlabel('Wine Class')
plt.ylabel('Count')
plt.show()


print("\n\n2.Descriptive Statistics:")
features = ['Alcohol', 'Malic.acid', 'Ash']

print("\n  a)Descriptive Statistics For each feature in the dataset:")

for feature in features:
    mean = data[feature].mean()
    median = data[feature].median()
    mode = data[feature].mode()[0]

    variance = data[feature].var()
    std_dev = data[feature].std()
    coiff_var = variance / (mean ** 2)
    iqr = data[feature].quantile(0.75) - data[feature].quantile(0.25)

    skewness = data[feature].skew()
    kurtosis = data[feature].kurtosis()

    print(f'\n    Feature: {feature}')
    print(f'      Mean: {mean:.2f}')
    print(f'      Median: {median:.2f}')
    print(f'      Mode: {mode:.2f}')
    print(f'      Variance: {variance:.2f}')
    print(f'      Standard Deviation: {std_dev:.2f}')
    print(f'      Coefficient of Variation: {coiff_var:.2f}')
    print(f'      Interquartile Range (IQR): {iqr:.2f}')
    print(f'      Skewness: {skewness:.2f}')
    print(f'      Kurtosis: {kurtosis:.2f}')

print("\n  b)Summary statistics for all numerical columns:")
for column in data.describe().columns:
    print(f"\n    Statistics for {column}:")
    for name, value in zip(data.describe().index, data.describe()[column]):
        print(f"      {name}: {value:.2f}")

print("\n\n3.Group-Based Analysis (by Wine Class):")
mean = data.groupby('Wine').mean()
print("\n  a.1)Mean values for each wine class:")
for wine_class, row in mean.iterrows():
    print(f"    Wine Class {wine_class}:")
    for feature, value in row.items():
        print(f"      {feature}: {value:.2f}")

median = data.groupby('Wine').median()
print("\n  a.2)Median values for each wine class:")
for wine_class, row in median.iterrows():
    print(f"    Wine Class {wine_class}:")
    for feature, value in row.items():
        print(f"      {feature}: {value:.2f}")

print("\n\n4.Data Visualization:")
print("\n  a) pair plot to visually explore the relationships between features (alcohol, malic acid, ash):")
sns.pairplot(data[features])
plt.suptitle('Pair Plot of Alcohol, Malic Acid, and Ash', y=1.02)
plt.show()

print("\n\n   b)Heatmap of the correlation matrix:")
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()

print("\n\n   c)Scatter Matrix of Alcohol, Malic Acid, and Ash with Class Separation:")
scatter_matrix(data[features + ['Wine']], c=data['Wine'], alpha=0.5, figsize=(13, 8), marker='o')
plt.suptitle('Scatter Matrix of Alcohol, Malic Acid, and Ash with Class Separation', y=1.02)
plt.show()

print("\n\n5. Correlation Analysis:")

cor_alc_ash = data['Alcohol'].corr(data['Ash'])
cor_mal_flav = data['Malic.acid'].corr(data['Flavanoids'])
print("\n  a)Pearson correlation coefficient for eachpair of features (alcohol vs. ash, malic acid vs. flavanoids):")
print(f"\n      Correlation coefficient between Alcohol and Ash is {cor_alc_ash}")
print(f"      Correlation coefficient between Malic Acid and Flavanoids is {cor_mal_flav}")

print("\n  b)Discussion of Significant Patterns:\n")
print("\n  Alcohol And Ash Content:\n")
if cor_alc_ash > 0.7:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a strong positive relationship between Alcohol and Ash, suggesting that wines with higher alcohol content tend to have higher ash content. This may be related to grape characteristics or winemaking processes.")
elif 0.3<=cor_alc_ash<=0.7:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a moderate positive relationship between Alcohol and Ash, indicating that wines with higher alcohol content may also have higher ash content.")
elif 0 < cor_alc_ash < 0.3:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a weak positive relationship between Alcohol and Ash, suggesting that other factors may influence these features.")
elif -0.3 < cor_alc_ash < 0:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a weak negative relationship between Alcohol and Ash, suggesting that other factors may influence these features.")
elif -0.7 < cor_alc_ash < -0.3:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a moderate negative relationship between Alcohol and Ash, suggesting that wines with higher alcohol content may have lower ash content.")
elif cor_alc_ash < -0.7:
    print(f"      Correlation coifficient {cor_alc_ash} tells us there is a strong negative relationship between Alcohol and Ash, indicating that wines with higher alcohol content tend to have lower ash content.")

print("\n  Malic Acid And Flavanoids:\n")
if cor_mal_flav > 0.7:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a strong positive relationship between Malic Acid and Flavanoids, suggesting that wines with higher malic acid levels tend to have higher flavonoid content.")
elif 0.3 < cor_mal_flav <= 0.7:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a moderate positive relationship between Malic Acid and Flavanoids, indicating that wines with higher malic acid levels may also have higher flavonoid content.")
elif 0 < cor_mal_flav < 0.3:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a weak positive relationship between Malic Acid and Flavanoids, suggesting that other factors may influence these features.")
elif -0.3 < cor_mal_flav < 0:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a weak negative relationship between Malic Acid and Flavanoids, suggesting that other factors may influence these features.")
elif -0.7 < cor_mal_flav < -0.3:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a moderate negative relationship between Malic Acid and Flavanoids, suggesting that wines with higher malic acid levels may have lower flavonoid content.")
elif cor_mal_flav < -0.7:
    print(f"      Correlation coifficient {cor_mal_flav} tells us there is a strong negative relationship between Malic Acid and Flavanoids, indicating that wines with higher malic acid levels tend to have lower flavonoid content.")
```
